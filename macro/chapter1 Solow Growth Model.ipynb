# section.1 Solow Growth Model

## 1.2 basic assumption


- continuous time
- A closed economy with no public spending.
- Households are assumed to save a constant exogenous fraction ,s ∈(0,1) of their disposable income $$c(t)=(1-s)Y(t)\tag {1}$$  where C (t ) is the total consumption and Y (t ) is the total income

- All firms have access to the same technology, and the aggregate production function is Labor-augmenting for convenience  $$Y(t)=F(K(t),A(t)L(t)) \tag {2}$$
    - Y (t ): the total amount of production of the final good at time t
    - K (t ): capital stock (equipments, lands, etc)
    - L(t ): total employment (hours of employment or number of employees)
    - A(t ): technology or effectiveness of labor
    - A(t )L(t ): effective labo

- Constant returns to scale: $F (cK ,cAL) = cF (K ,AL)$
- Output per unit of effective labor: $$y=\frac{Y}{AL}=\frac{F(K,AL)}{AL}=F(\frac{K}{AL},1)=F(k,1)=f(k)$$
    - $k=\frac{K}{AL}$
    - $f'(k)>0;f''(k)<0;f(0)=0$
    - $f'(k) = \frac{\partial F}{\partial K}=MPK$  (Marginal Product of Capital)
    - inada conditions(稻田条件):$$\lim_{k \to 0} f'(k)=\infty ; \lim_{k \to \infty} f'(k)=0$$
        - 稻田条件保证了：存在唯一的 $k^* > 0$ 使得 $\dot{k} = 0$ 该稳定状态是全局稳定的


## 1.3 intensive form （集约形式）


- 什么是集约形式，以及为什么叫集约形式（from deepseek）
    - 集约形式 指的是将宏观总量生产函数，转化为人均或单位有效劳动形式的过程。
    - 这个名字源于“集约型增长”和“粗放型增长”的区分。
        - 粗放型增长：指依靠增加生产要素的数量（投入更多的资本 K 和劳动 L）来推动经济增长。
        - 对应形式：生产函数的原始形式 Y = F(K, L) 直接体现了这种关系。要增加总产出 Y，你就需要增加总投入 K 和 L。

        - 集约型增长：指依靠提高生产要素的使用效率来推动经济增长。即在人均或单位投入的层面上获得更多产出。
        - 对应形式：生产函数的集约形式 y = f(k) 完美地刻画了这一点。它关注的不再是总规模，而是人均资本 k 的积累和技术水平 如何决定了人均产出 y 的效率。

    - “集约形式”之所以叫这个名字，是因为它将分析问题的视角从“总量”和“规模”（粗放）转向了“人均”和“效率”（集约）。它让我们能够集中分析在剔除了经济规模大小的影响后，一个经济的**内在效率**和**发展水平**（由 k 和 A 代表）是如何决定其人均产出的。



### The rental price of capital is
$$r_K=MPK=\frac{\partial F(K,AL)}{\partial K}=f'(k)\tag {3}$$
### Note that the interest rate is the rental price minus the depreciation rate of capital
$$r=r_K-\delta=f'(k)-\delta$$

### wage:
$$W=MPL=\frac{\partial F(K,AL)}{\partial L}=A[f(k)-f'(k)k]\tag {4}$$
MPL:Marginal Product of Labor



- 一些补充
    - 对于$W=MPL$（该式子代表工资率的分配）：在竞争均衡中，工资率等于劳动的边际产品，即W=MPL。因此，公式表明工资率取决于技术进步水平A和单位有效劳动的产出结构。
    - 对于$MPL=A[f(k)-f'(k)k]$&$MPK=f'(k)$:反映了产出在劳动和资本之间的分配。
        - $\text{资本总收入} = \text{MPK} \times K = f'(k) \times K$

        - $\text{劳动总收入} = \text{MPL} \times L = A[f(k) - f'(k)k] \times L$

        - $\frac{\text{劳动总收入}}{Y} = \frac{A[f(k) - f'(k)k] \times L}{AL \cdot f(k)} = \frac{f(k) - f'(k)k}{f(k)} = 1 - \frac{f'(k)k}{f(k)}$

        - $\frac{\text{资本总收入}}{Y} = \frac{f'(k) \times K}{AL \cdot f(k)} = \frac{f'(k) \times (kAL)}{AL \cdot f(k)} = \frac{f'(k)k}{f(k)}$

        - 柯布道格拉斯形式下的份额分配为？？？


    - 欧拉定理验证：
        - 对于一次齐次函数（规模报酬不变），欧拉定理指出：$$F(K, AL) = \frac{\partial F}{\partial K} \cdot K + \frac{\partial F}{\partial (AL)} \cdot (AL)$$

        - 计算对资本偏导数：$\frac{\partial Y}{\partial K} = \frac{\partial}{\partial K}[AL \cdot f(k)] = AL \cdot f'(k) \cdot \frac{\partial k}{\partial K}$ 

        由于 $k = \frac{K}{AL}$，所以 $\frac{\partial k}{\partial K} = \frac{1}{AL}$，

        因此：$\frac{\partial Y}{\partial K} = AL \cdot f'(k) \cdot \frac{1}{AL} = f'(k)$

        - 计算对有效劳动偏导数：$\frac{\partial Y}{\partial (AL)} = \frac{\partial}{\partial (AL)}[AL \cdot f(k)] = f(k) + AL \cdot f'(k) \cdot \frac{\partial k}{\partial (AL)}$

        由于 $k = \frac{K}{AL}$，所以 $\frac{\partial k}{\partial (AL)} = -\frac{K}{(AL)^2} = -\frac{k}{AL}$，因此

        $\frac{\partial Y}{\partial (AL)} = f(k) + AL \cdot f'(k) \cdot \left(-\frac{k}{AL}\right) = f(k) - f'(k)k$

        - 验证欧拉定理 $\frac{\partial Y}{\partial K} \cdot K + \frac{\partial Y}{\partial (AL)} \cdot (AL) = f'(k) \cdot K + [f(k) - f'(k)k] \cdot (AL)$

        代入 $K = k \cdot AL$：

         $= f'(k) \cdot (kAL) + [f(k) - f'(k)k] \cdot (AL)$

         $= AL[f'(k)k + f(k) - f'(k)k] = AL \cdot f(k) = Y$

$$\frac{\dot L}{L}=n   ;  \frac{\dot A}{A}=g$$
Labor knowledge grow are constant rates

$$\dot{K}(t) = sY(t) - \delta K(t) \tag {5}$$

- $\dot{K}(t)$ - 资本积累：资本存量随时间的变化率
- sY(t) - 投资：储蓄率 s 乘以总产出 Y(t)，代表新增资本形成
- $\delta K(t)$ - 折旧：以 δ 的速率折旧的资本存量
- s - 储蓄率：收入中用于储蓄的固定比例，其中 s ∈ (0,1)
- Y(t) - 总产出：时间 t 时经济体的总产量
- δ - 折旧率：资本磨损或过时的速率,$\delta \in (0,1)$
- 当 $\dot {K}(t) = 0$ 达到稳态



在索洛模型中，"路径"指的是变量随时间变化的轨迹或过程。
- A(t)的路径：技术水平随时间变化的轨迹
    通常假设为指数增长：$A(t) = A(0)e^{gt}$，其中 $g$ 是外生技术进步率

- L(t)的路径：劳动力随时间变化的轨迹
    通常假设为指数增长：$L(t) = L(0)e^{nt}$其中 $n$ 是外生人口增长率

In the basic Solow model for a given path of A(t ) and L(t ); and an initialstock of K (0), an equilibrium path is a path of the following endogenousvariables: **capital stocks, output levels, consumption levels, wages, and rental rates** ，以上变量参考公式（1）~（5）

The condition  $\dot K(t)=0$ CANNOT imply a steady state for
$K (t)$
- 稳态要求$k$为0，而不是$K$

The key equation of the Solow model
$$\dot {k}(t) = sf (k(t)) −(n + g + \delta )k(t)$$

上式推导过程为$k$对$t$求导，此处省略
$$k(t)=\frac {K(t)}{A(t)L(t)}$$

Graphic Explanation of the Key Equation（一个图例）

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

# 定义参数
alpha = 0.3      # 资本份额
s = 0.3          # 储蓄率
n = 0.01         # 人口增长率
g = 0.02         # 技术进步率
delta = 0.05     # 折旧率

# 创建资本存量k的范围
k = np.linspace(0.1, 10, 100)

# 定义柯布-道格拉斯生产函数
def f(k, alpha=alpha):
    return k ** alpha

# 计算各项投资
actual_investment = s * f(k)          # 实际投资 sf(k)
break_even_investment = (n + g + delta) * k  # 持平投资 (n+g+δ)k

# 计算稳态资本存量 k*
# 在稳态时：s * k^α = (n+g+δ) * k
k_star = (s / (n + g + delta)) ** (1 / (1 - alpha))

# 创建图形
plt.figure(figsize=(10, 8))

# 绘制投资曲线
plt.plot(k, actual_investment, 'b-', linewidth=2, label='实际投资 $sf(k)$')
plt.plot(k, break_even_investment, 'r-', linewidth=2, label='持平投资 $(n + g + \\delta)k$')

# 标记稳态点
plt.axvline(x=k_star, color='gray', linestyle='--', alpha=0.7)
plt.axhline(y=s * f(k_star), color='gray', linestyle='--', alpha=0.7)

# 标记稳态点
plt.plot(k_star, s * f(k_star), 'ro', markersize=8)
plt.annotate(f'$k^*$ = {k_star:.2f}', 
             xy=(k_star, 0), 
             xytext=(k_star, -0.02),
             ha='center', 
             fontsize=12)

# 设置坐标轴范围
plt.xlim(0, 10)
plt.ylim(0, 0.8)

# 添加标签和标题
plt.xlabel('单位有效劳动资本 $k$', fontsize=14)
plt.ylabel('单位有效劳动投资', fontsize=14)
plt.title('索洛模型：投资与资本积累', fontsize=16)

# 添加图例
plt.legend(fontsize=12)

# 添加网格
plt.grid(True, alpha=0.3)

# 在图上添加文本说明
plt.text(7, 0.2, '稳态条件：\n$sf(k^*) = (n+g+\\delta)k^*$', 
         fontsize=12, bbox=dict(boxstyle="round,pad=0.3", facecolor="white"))

# 显示图形
plt.tight_layout()
plt.show()

# 打印稳态值信息
print(f"稳态资本存量 k* = {k_star:.4f}")
print(f"稳态产出 f(k*) = {f(k_star):.4f}")
print(f"稳态实际投资 sf(k*) = {s * f(k_star):.4f}")
print(f"稳态持平投资 (n+g+δ)k* = {break_even_investment[np.argmin(np.abs(k - k_star))]:.4f}")
```

# section.2 Balenced Growth Path

the BGP represents a situation where each variable of the
model is growing at a constant rate
BGP: Y , C , K grow at a constant rates

## 2.1 平衡增长路径上的增长率



### 总资本增长率

$$g_K^{BGP} = \frac{\dot{K}}{K}\bigg|_{BGP} = \frac{\dot{k}}{k}\bigg|_{k=k^*} + \frac{\dot{A}}{A} + \frac{\dot{L}}{L} = n + g
$$

### 总产出增长率

$$g_Y^{BGP} = \frac{\dot{Y}}{Y}\bigg|_{BGP} = \frac{\dot{y}}{y}\bigg|_{y=f(k^*)} + \frac{\dot{A}}{A} + \frac{\dot{L}}{L} = n + g$$


### 总消费增长率

$$g_C^{BGP} = \frac{\dot{C}}{C}\bigg|_{BGP} = \frac{\dot{Y}}{Y}\bigg|_{BGP} = n + g$$


### 人均产出增长率

$$g_{Y/L}^{BGP} = g_Y^{BGP} - g_L = g$$

稳态时，$\dot K = 0,\dot C = 0,\dot Y = 0$

- Can you write $g_{\frac{Y}{L}}$ as a function of $k(t)$ and  $\dot k(t)$ if the economy has not reached on BGP yet
  - $y=\frac{Y}{L}=Af(k)$

  - $g_{\frac{Y}{L}}=\frac{dy/dt}{y}$
  - 其中 $$\frac {dy}{dt}=\frac {d}{dt}[Af(k)]=\frac {dA}{dt}f(k)+Af'(k)\frac {dk}{dt} $$
  - 假设技术进步率恒定 $\frac{dA}{dt}=gA$,且 $\frac {dk}{dt} =\dot k(t)$
  - 代入得：$$\frac{dy}{dt}=gAf(k)+Af'(k)\dot k(t)$$ 
  - 因此 $$g_{\frac {Y}{L}}=g+\frac {f'(k)}{f{k}} \dot k(t)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

# 定义参数
alpha = 0.3      # 资本份额
n = 0.01         # 人口增长率
g = 0.02         # 技术进步率
delta = 0.05     # 折旧率
s_old = 0.2      # 旧储蓄率
s_new = 0.3      # 新储蓄率

# 创建资本存量k的范围
k = np.linspace(0.1, 8, 100)

# 定义柯布-道格拉斯生产函数
def f(k, alpha=alpha):
    return k ** alpha

# 计算各项投资
break_even_investment = (n + g + delta) * k    # 持平投资 (n+g+δ)k
old_actual_investment = s_old * f(k)           # 旧实际投资 s_OLD f(k)
new_actual_investment = s_new * f(k)           # 新实际投资 s_NEW f(k)

# 计算新旧稳态资本存量
k_old_star = (s_old / (n + g + delta)) ** (1 / (1 - alpha))
k_new_star = (s_new / (n + g + delta)) ** (1 / (1 - alpha))

# 创建图形
plt.figure(figsize=(12, 8))

# 绘制投资曲线
plt.plot(k, break_even_investment, 'r-', linewidth=3, label='持平投资 $(n + g + \\delta)k$')
plt.plot(k, old_actual_investment, 'b--', linewidth=2, label='旧实际投资 $s_{OLD}f(k)$')
plt.plot(k, new_actual_investment, 'b-', linewidth=2, label='新实际投资 $s_{NEW}f(k)$')

# 标记新旧稳态点
plt.axvline(x=k_old_star, color='gray', linestyle=':', alpha=0.7)
plt.axvline(x=k_new_star, color='gray', linestyle=':', alpha=0.7)

# 标记稳态点
plt.plot(k_old_star, s_old * f(k_old_star), 'ro', markersize=8)
plt.plot(k_new_star, s_new * f(k_new_star), 'ro', markersize=8)

# 添加稳态标注
plt.annotate(f'$k_{{OLD}}^*$', 
             xy=(k_old_star, 0), 
             xytext=(k_old_star, -0.02),
             ha='center', 
             fontsize=14)

plt.annotate(f'$k_{{NEW}}^*$', 
             xy=(k_new_star, 0), 
             xytext=(k_new_star, -0.02),
             ha='center', 
             fontsize=14)

# 添加箭头表示变动方向
plt.annotate('', 
             xy=(k_new_star, s_new * f(k_new_star)/2), 
             xytext=(k_old_star, s_old * f(k_old_star)/2),
             arrowprops=dict(arrowstyle='->', color='green', lw=2))

plt.text((k_old_star + k_new_star)/2, s_new * f(k_new_star)/3, 
         '储蓄率提高\n稳态资本增加', 
         ha='center', va='center',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"),
         fontsize=12)

# 设置坐标轴范围
plt.xlim(0, 8)
plt.ylim(0, 0.6)

# 添加标签和标题
plt.xlabel('单位有效劳动资本 $k$', fontsize=14)
plt.ylabel('单位有效劳动投资', fontsize=14)
plt.title('索洛模型：储蓄率提高对稳态资本的影响', fontsize=16)

# 添加图例
plt.legend(fontsize=12, loc='upper left')

# 添加网格
plt.grid(True, alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()

# 打印新旧稳态值
print(f"旧稳态资本存量 k_OLD* = {k_old_star:.4f}")
print(f"新稳态资本存量 k_NEW* = {k_new_star:.4f}")
print(f"稳态资本增加量 = {k_new_star - k_old_star:.4f}")
print(f"增长率 = {(k_new_star - k_old_star) / k_old_star * 100:.2f}%")
```

### 2.2.1 Impact on the BGP & Growth Rate


At the BGP,$sf(k^*(s)) \equiv (n+g+\delta)k^*(s)$

*这里$k$是对$s$的函数，而不是$t$

等式两边对 $s$ 求导:$$f(k^*(s)) + sf'(k^*(s)) \frac{\partial k^*}{\partial s} = (n + g + \delta) \frac{\partial k^*}{\partial s}$$
$\Rightarrow$
$$\frac{\partial k^*}{\partial s} = \frac{f(k^*)}{(n + g + \delta) - sf'(k^*)} > 0$$

or

$$\frac{\partial k^*}{\partial s} = \frac{f(k^*)k^*}{s[f(k^*) - f'(k^*)k^*]} > 0$$

$\dot k(t)=sf(k(t))-(n+g+\delta)k(t)$ jumps up from zero initially and
will eventually fall back to zero.

A change in the saving rate has a level effect but not a growth effect
- level effect：人均产出**水平**的增加
- growth effect：人均产出**增长率**的增加，这个要靠技术进步。

### 2.2.3 Impact on consumption



消费水平：$$c=(1-s)f(k)$$

In the short run, it will jump down, and then start to grow.We need to analyze further: what is the relationship between savingrate and consumption on the BGP

- Golden Rule

稳态时有：$$sf(k^*)=(n+g+\delta)k^*$$
稳态的消费：$$c^*=(1-s)f(k^*(s))=f(k^*(s))-(n+g+\delta)(k^*(s))$$
c对s求偏导：$$\frac{\partial c^*}{\partial s}={f'(k^*(s))-(n+g+\delta)} \frac{\partial k^*(s)}{\partial s}$$

- “Golden Rule” capital level is a specific steady state for a particular saving rate where the consumption is at its maximum possible level on the BGP



### 2.2.4 The Effect on Output in the Long Run


the effect of change of saving rate on output:$$\frac {\partial y^*}{\partial s}=f'(k^*)\frac {\partial k^*}{\partial s}=\frac {\partial {f'(k^*)f(k^*)}}{\partial (n+g+\delta)-sf'(l^*)}$$


elasticity of output with respect to the saving rate at BGP
$$E(y^*,s)=\frac {\partial y^*}{\partial s}\frac {s}{y^*}=\frac {\alpha_k (k^*)}{1-\alpha_k(k^*)}$$
where $\alpha_k(k^*)=\frac {k^*f'(k^*)}{f(k^*)}$ is the elasticity of output with respect to capital at BGP,这里的$\alpha_k$也表示资本份额

- 对于柯布-道格拉斯生产函数 $f(k)=k^\alpha$，可以推导出 $E(y^*,s)=\frac {\alpha }{1-\alpha}$


### 2.2.5 The Speed of Convergence (k)
Taylor expansion (linearizing) of $\dot k (k)$ around $k^*$
 $$\dot{k} \simeq \dot{k}(k^*) + \left[ \frac{\partial \dot{k}(k)}{\partial k} \right]_{k=k^*} (k-k^*) = \left[ \frac{\partial \dot{k}(k)}{\partial k} \right]_{k=k^*} (k-k^*)\simeq  -\lambda(k(t)-k^*) $$

- $\simeq$:近似等于approximately equal to

- 收敛速度 $\lambda$ 定义为资本存量向稳态调整的速率 or the growth rate of $k^*-k(t)$ : $$k(t)-k^*\simeq e^{-\lambda t}[k(0)-k^*]$$

为什么近似等于是泰勒展开形式
- 索洛模型的基本方程是非线性的,非线性方程通常难以直接求解，而泰勒展开可以将非线性系统在稳态点附近线性化，从而简化分析。
- 解微分方程$$\frac{d(k-k^*)}{dt}=\dot k \approx -\lambda(k-k^*)$$可得到上式



### 2.2.6speed of convergence of $y(t)$ around $y^*$,


$$\dot y=f'(k)\dot k$$


$$ \dot{y} \simeq \left[ \frac{\partial \dot{y}}{\partial y} \right]_{y=y^*} (y - y^*) = \left[ \frac{\partial \dot{y}}{\partial k} \right]_{k=k^*} \left[ \frac{\partial k}{\partial y} \right]_{y=y^*} (y - y^*) $$


$$\simeq \left[ \frac {\partial f'(k)k}{\partial k} \cdot \frac{1}{f'(k)} \right]_{k=k^*} (y - y^*) \quad \text{(注: } \partial k / \partial y = 1/f'(k) \text{)} $$


$$  \simeq \left[ \frac{f''(k)\dot k}{f'(k)} + \frac{\partial \dot{k}(k)}{\partial k} \right]_{k=k^*} (y - y^*) $$


$$ \simeq \left[ \frac{\partial \dot{k}(k)}{\partial k} \right]_{k=k^*} (y - y^*) \simeq -\lambda [y(t) - y^*] $$



### 2.2.7 Find $\lambda$


$$
\lambda = -\frac{\partial \dot{k}(k)}{\partial k} \bigg|_{k=k^*}
$$

$$
= -[sf'(k^*) - (n + g + \delta)]
$$

$$
= (n + g + \delta) - \frac{(n + g + \delta)k^*f'(k^*)}{f(k^*)} \quad (*)
$$

= [1 - \alpha_k(k^*)](n + g + \delta).

## 2.3 The Soolow Residual


### 2.3.1 增长率分解

- 生产函数$$Y(t)=F[K(t),A(t)L(t)]$$
- 增长率分解 $$\frac {\dot Y(t)}{Y(t)}=\alpha_K (t)\frac {\dot K}{K}+\alpha_L (t)\frac {\dot L}{L}+R(t)$$
    -  $\alpha_{K/L} (t)$ 资本/劳动产出弹性
    - $R(t)$ 索洛剩余，这是公式中无法被资本和劳动增长所解释的那部分产出增长。它被解释为技术进步的速率，即全要素生产率（TFP）的增长率。它捕捉了所有除资本和劳动数量增加之外的效率提升。
    - where   $\alpha K (t) + \alpha L(t)$= 1

- 人均产出的增长率分解 $$\frac {\dot Y(t)}{Y(t)}-\frac {\dot L(t)}{L(t)}=\alpha_K (t) [ \frac {\dot K}{K}-\frac {\dot L}{L} ] +R(t)$$


### 2.3.2 扩展的索洛增长模型


$$Y(t)=K(t)^\alpha R(t)^\beta T(t)^\gamma[A(t)L(t)]^{1-\alpha-\beta-\gamma}$$

- $K(t)$（资本）： 可积累的生产资料。
- $R(t)$（自然资源）： 如石油、矿产等，其存量会随着使用而消耗。
- $T(t)$（土地）： 固定的生产要素，总量不变。
- $A(t)L(t)$（有效劳动）： 经过技术进步增强的劳动力。
所有要素的指数之和为1。这表示该生产函数是规模报酬不变的——如果所有要素同时翻倍，产出也会翻倍。
-  $\dot R(t)=-bR(t),b>0$  自然资源存量以一个固定的速率b 衰减（例如，每年减少1%）。这表明资源是不可再生的，正在被持续开采和消耗。
- 土地固定: $\dot T(t)=0$,土地的数量是固定不变的，不会增长。

基本增长核算方程
$$g_y(t) = \alpha g_k(t) + \beta g_r(t) + \gamma g_t(t) + (1-\alpha-\beta-\gamma)[g_a(t) + g_l(t)]$$
$$ = \alpha g_k(t) - \beta b + (1-\alpha-\beta-\gamma)(n + g)$$


总产出增长率
$$
g_{Y}^{bgp} = \frac{(1-\alpha-\beta-\gamma)(n+g)-\beta b}{1-\alpha}
$$

人均产出增长率  
$$
g_{Y/L}^{bgp} = \frac{(1-\alpha-\beta-\gamma)g-\beta b-(\beta+\gamma)n}{1-\alpha}
$$

$g_Y(t)$：瞬时总产出增长率 - 在任何时点t的总产出变化率

$g_Y^{bgp}$：平衡增长路径上的总产出增长率 - 经济收敛后的长期稳定增长率
与传统索洛模型相比，长期增长率受到资源约束的影响.经济从任意初始状态出发，其瞬时增长率 $g_Y(t)$ 会逐渐收敛到平衡增长路径的增长率 $g_Y^{bgp}$

- **传统模型增长率**: $g$
- **扩展模型增长率**: $g - \frac{\beta}{1-\alpha}b - \frac{\gamma}{1-\alpha}n$
- 只要生产函数满足新古典假设（规模报酬不变、要素边际产出递减），经济仍然会从任何初始资本存量水平收敛到一条稳定的增长路径。
