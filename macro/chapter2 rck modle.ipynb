# introduction RCK model

-  It is one of models that resemble the Solow model but in which the saving rate is endogenous.
- Saving is determined by the consumers’ optimal choices between current and future consumptions.
- The RCK model represents a market competitive equilibrium with microfoundation.
- The equilibrium path of RCK model are paths of endogenous variables
 (such as y(t),c(t),k(t),r(t),w(t)) satisfying
  - Firms’ profit maximization for each t;
    - $$\pi =F(K,L)-rK-wL$$
    - 资本的边际产出等于利率，劳动的边际产出等于工资，只要你知道了$ k(t)$，你就能立刻求出$r(t)$ 和 $w(t)$
  - Households’ utility maximization over lifetime;
    - 家庭的效用来自消费 c(t)。他们通过将收入（来自工资和储蓄利息）在消费和储蓄之间进行分配来做决策。欧拉方程描述了家庭如何平滑消费，它连接了消费的增长率与利率 .
     $$\frac{\dot c(t)}{c(t)}=(1-\theta)[r(t)-\rho]$$

    - 其中  $\dot c$ 是消费的导数（变化量）， $\theta$ 是相对风险厌恶系数（衡量消费平滑的意愿）， $\rho$ 是时间偏好率（衡量耐心程度）。
  - Market clearing for each t
